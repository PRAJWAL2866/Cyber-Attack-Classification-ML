```python
# RQ7: Final Decision Matrix

import pandas as pd
import matplotlib.pyplot as plt


# USE YOUR ACTUAL RESULTS (adjust values if needed)

data = {
    "Model": ["Logistic", "Random Forest", "SVM"],
    
    # From RQ2 / RQ5
    "Performance": [0.66, 0.63, 0.66],  
    
    # Interpretability (manually assigned but logical)
    "Interpretability": [0.9, 0.7, 0.6],
    
    # From RQ6 (average CV score approx)
    "Robustness": [0.64, 0.63, 0.64]
}

df = pd.DataFrame(data)

print(df)

df.to_csv("RQ7_table.csv", index=False)

# PLOT (STACKED BAR)

ax = df.set_index("Model").plot(kind="bar", figsize=(8,5))

# Add value labels
for container in ax.containers:
    ax.bar_label(container, fmt="%.2f", padding=2)

plt.title("Final Decision Matrix")
plt.ylabel("Score")

plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig("RQ7_figure.pdf")
plt.show()
```
